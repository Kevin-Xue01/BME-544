## BME 544 DC 4

```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from file
data = np.loadtxt('DC_4/Different Objects.txt', delimiter='\t', skiprows=1)
time = data[:, 0]
signal = data[:, 1]

# Plot
plt.figure(figsize=(22, 5))
plt.plot(time, signal, linestyle='-', color='b', label='Signal')
plt.axvline(x=3.8, alpha=0.5)
plt.axvline(x=10.8, alpha=0.5)
plt.axvline(x=16, alpha=0.5)
plt.axvline(x=24.5, alpha=0.5)

plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.title('Time vs Signal')
plt.legend()
plt.grid(True)
plt.show()
```
